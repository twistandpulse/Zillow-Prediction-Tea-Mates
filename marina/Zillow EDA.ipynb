```python
import pandas as pd
```

```python
zri = pd.read_csv('ZRI.csv')
zri
```

```python
zri.columns
```

```python
zri = zri.drop(columns=['RegionID', 'RegionName','City', 'State', 'Metro', 'CountyName','SizeRank'])
```

```python
zri['missing'] = zri.isnull().sum(axis=1)
```

```python
zri = zri[zri['missing']<50]
```

```python
zri = zri.fillna(method = 'ffill', axis=1, limit=100)
```

```python
zri['missing'] = zri.isnull().sum(axis=1)
zri['missing'].sort_values(ascending=False)
```

```python
zri = zri.fillna(method = 'backfill', axis=1, limit=100)
```

```python
zri['missing'] = zri.isnull().sum(axis=1)
zri['missing'].sort_values(ascending=False)
```

```python
zri.to_csv('copy.csv')
```

```python
zri_copy = pd.read_csv('ZRI.csv')
zri_copy = zri_copy.drop(columns=['City', 'State', 'Metro', 'CountyName','SizeRank'])
zri_copy['missing'] = zri_copy.isnull().sum(axis=1)
zri_copy = zri_copy[zri_copy['missing']<50]
zri_copy = zri_copy[['RegionID', 'RegionName']]
zri_copy
```

```python
zri = pd.concat([zri_copy,zri],axis=1)
zri
```

```python
zri['2011'] = zri.iloc[:,6:18].mean(axis=1)
zri['2012'] = zri.iloc[:,18:30].mean(axis=1)
zri['2013'] = zri.iloc[:,30:42].mean(axis=1)
zri['2014'] = zri.iloc[:,42:54].mean(axis=1)
zri['2015'] = zri.iloc[:,54:66].mean(axis=1)
zri['2016'] = zri.iloc[:,66:78].mean(axis=1)
zri['2017'] = zri.iloc[:,78:90].mean(axis=1)
zri['2018'] = zri.iloc[:,90:102].mean(axis=1)
zri['2019'] = zri.iloc[:,102:114].mean(axis=1)
```

```python
zri
```

```python
zri_2011 = zri[['RegionName','2011']]
zri_2012 = zri[['RegionName','2012']]
zri_2013 = zri[['RegionName','2013']]
zri_2014 = zri[['RegionName','2014']]
zri_2015 = zri[['RegionName','2015']]
zri_2016 = zri[['RegionName','2016']]
zri_2017 = zri[['RegionName','2017']]
zri_2018 = zri[['RegionName','2018']]
zri_2019 = zri[['RegionName','2019']]
```

```python
zri_2011.rename(columns={'2011':'zri'}, inplace=True)
zri_2012.rename(columns={'2012':'zri'}, inplace=True)
zri_2013.rename(columns={'2013':'zri'}, inplace=True)
zri_2014.rename(columns={'2014':'zri'}, inplace=True)
zri_2015.rename(columns={'2015':'zri'}, inplace=True)
zri_2016.rename(columns={'2016':'zri'}, inplace=True)
zri_2017.rename(columns={'2017':'zri'}, inplace=True)
```

```python
zri_2011.loc[:,'year'] = '2011'
zri_2012.loc[:,'year'] = '2012'
zri_2013.loc[:,'year'] = '2013'
zri_2014.loc[:,'year'] = '2014'
zri_2015.loc[:,'year'] = '2015'
zri_2016.loc[:,'year'] = '2016'
zri_2017.loc[:,'year'] = '2017'
zri_2018.loc[:,'year'] = '2018'
zri_2019.loc[:,'year'] = '2019'
```
